```python
# Сначала следует понять насколько много пропущенных значений по сравнению с размером таблицы данных
data_rows_total = data.shape[0]    # Определение общего количества строк в исходной таблице и запись в переменную data_rows_total
print('Общее количество строк в исходной таблице составляет:', data_rows_total)   # Вывод на экран переменной data_rows_total

data_rows_na = data['total_income'].isna().sum()    # Определение количества пропущенных значений в столбце total_income и запись в переменную
print('Общее количество пропущенных значений в столбце "ежемесячный доход" составляет:', data_rows_na)   # Вывод на экран переменной data_rows_total

try:    # Использование конструкции try-except чтобы защититься от деления на ноль
    data_percent_na = data_rows_na / data_rows_total    # Вычисление отношения количества пропущенных значений к общему количеству строк таблицы
    print('Число пропущенных значений в исходной таблице составляет {:.0%} от общего количества данных'.format(data_percent_na))
    if data_percent_na > 0.1:
        print('Вывод: число пропущенных значений составляет более 1%, следовательно, удалить строки с пропущенными значениями нельзя, их следует заполнить и использовать в анализе')
    elif data_percent_na == 0:
        print('Вывод: в таблице данных нет пропущенных значений, либо пропуски уже были заполнены')
    else:
        print('Вывод: число пропущенных значений составляет менее 1%, следовательно, этими данными можно пренебречь и удалить их')
except:
    print('Ошибка при расчёте, убедитесь, что данные были импортированы корректно')
    

# Заполнение пропущенных значений
# Поскольку пропущенные значения столбца days_employed в решении поставленных в проекте задач ценности не имеют, допустимо заполнить их нулями
data['days_employed'] = data['days_employed'].fillna(0)    # Применение метода fillna с аргументом 0 к столбцу days_employed

# Поскольку нельзя определить ежемесячный доход для пропущенных значений, их следует заполнить условным значением, например, -1,
# и в дальнейшем анализе по столбцу total_income их не учитывать, а в анализе по другим столбцам - учитывать.
data['total_income'] = data['total_income'].fillna(-1)    #  Применение метода fillna с аргументом -1 к столбцу total_income

# Проверка результатов заполнения пропусков
print()
print('Проверка результатов заполнения пропусков:')
data.info()
```

1. Выделить леммы в значениях столбца с целями получения кредита: описать, как проводилась лемматизация целей кредита;

```python
from collections import Counter    # Импорт коллекции Counter
from pymystem3 import Mystem    # Импорт библиотеки Mystem
m = Mystem()    # переменной m присвоить Mystem() 


# Выделим список уникальных значений строк в столбце purpose и выполним с ним лемматизацию
data_purpose_unique = data['purpose'].unique()    # Выявление уникальных строк в столбце purpose с последующей записи результата в переменную data_purpose_unique
lemmas = m.lemmatize(' '.join(data_purpose_unique))    # Выполнение леммматизации функцией lemmatize() с аргументом в виде строки, полученной методом join() с аргументом в виде переменной data_purpose_unique
print(Counter(lemmas))    # Подсчёт количества повторяющихся лемм с помошью функции Counter() и вывод результата на экран
print()

# Исходя из обнаруженных наиболее часто встречающихся слов в столбце purpose можно определить категории целей получения кредита
def purpose_lemmatize(row):    # Функция которая возвращает наименование категории в зависимости от найденных в исходном столбце слов
    lemmas = m.lemmatize(row)    # Привести к лемме строковое значение столбца
    if 'недвижимость' in lemmas or 'жилье' in lemmas or 'строительство' in lemmas:   # Если встречаются перечисленные слова, то вернуть название соответствующей им категории
        return 'недвижимость'
    elif 'автомобиль' in lemmas:
        return 'автомобиль'
    elif 'образование' in lemmas:
        return 'образование'
    else:    # Наименее часто встречающиеся цели кредита (свадьба, ремонт) попадают в категорию "прочее"
        return 'прочее'

# Применим написанную функцию к столбцу purpose. Выполнение функции занимает примерно 3-4 секунды
data['purpose_category'] = data['purpose'].apply(purpose_lemmatize)     # С помощью метода apply() применяем функцию
#purpose_lemmatize к столбцу purpose, а результат записываем в новый столбец purpose_category
 
# Выполним подсчёт количества клиентов по категориям целей кредита
print(Counter(data['purpose_category']))



# Пробовал стемминг. Результат не устроил, поэтому оставил в комментариях.
#Стемминг дал неточный результат и операция занимает много больше времени по сравнению с лемматизацией и методом apply
#from nltk.stem import SnowballStemmer 
#russian_stemmer = SnowballStemmer('russian')
#for i in range(data.shape[0]):
#    for word in data.loc[i, 'purpose'].split():
#        stemmed_word = russian_stemmer.stem(word)
#        if stemmed_word == 'недвижим' or stemmed_word == 'жил' or stemmed_word == 'строительств':    # жил    автомобил   строительств
#            data.loc[i, 'purpose'] = 'недвижимость'
#        elif stemmed_word == 'автомобил':
#            data.loc[i, 'purpose'] = 'автомобиль'
#        elif stemmed_word == 'образован':
#            data.loc[i, 'purpose'] = 'образование'
#        else:
#            data.loc[i, 'purpose'] = 'прочее'
#Out:     'недвижимость': 9522, 'прочее': 4091, 'образование': 4013, 'автомобиль': 3828})
```

### Категоризация данных

- Категоризировать данные: перечислить, какие «словари» выделили для этого набора данных, и объяснить, почему.

Категоризовать данные по числу детей необязательно. Разбиение есть уже в исходных данных. Все-таки разница в возврате кредита у семей с одним ребенком или тремя детьми будет.


```python
# В начале описаны функции которые будут задействованы при категоризации
def children_cat(row):  # Функция для категоризации клиентов по количеству детей
    children = row
    if children == 0:     # Если количество детей равно нулю, значит нет детей
        return 'нет детей'
    elif 1 <= children <= 3:   # Если детей от 1 до 3 включительно будем считать что дети есть
        return 'есть дети'
    elif 4 <= children <= 10:    # Если детей от 4  до 10 включительно будем считать многодетной семьей
        return 'многодетная семья'
    return 'неверные данные'    # Во всех остальных случаях будем считать что данные указаны неверно


def total_income_cat(row):    # Функция для категоризации клиентов по ежемесячному доходу
    total_income = row
    if 0 < total_income <= 100000:    # Если ежемесячный доход в диапазоне от 0 до 100 тыс. то попадает с эту категорию
        return 'От 0 до 100 тыс.'
    elif 100000 < total_income <= 140000:
        return 'От 100 до 140 тыс.'
    elif 140000 < total_income <= 190000:
        return 'От 140 до 190 тыс.'
    elif 190000 < total_income <= 330000:
        return 'От 190 до 330 тыс.'
    elif total_income > 330000:
        return 'От 330 тыс. и выше'
    return 'Пропущенные данные'

    
# 1. Категоризация по столбцу "количество детей"
# Исследуем какое количество детей встречается в данных
print('В столбце "количество детей" встречаются следующие  значения:')
print(Counter(data['children']))    # Применение функции Counter для подсчета количества различных значений столбца children
print('Обнаружены выбросы: 20: 76, -1: 47, что составляет соответственно 0,35% и ')
print('0,22% от количества строк в таблице. Выбросы будут помещены в категорию "неверные данные"')
print()

# Категоризируем данные по столбцу "количество детей" и запишем результат в новый столбец 
data['children_category'] = data['children'].apply(children_cat)    # Применение метода apply() к столбцу children с функцией children_cat

# Выведем суммарные значения по новому столбцу children_category на экран для оценки
print('Результат работы функции категоризации по количеству детей:')
print(data['children_category'].value_counts())
print()

# 2. Категоризация по столбцу "ежемесячный доход"
# Исследуем какие суммы доходов и как часто встречаются в данных с помощью выборки из DataFrame с условием и методом count()
print('В столбце "ежемесячный доход" можно выделить следующие категории сумм которые встречаются с примерно одинаковой частотой:')
print('От 0 до 100 тыс.:', data[(data['total_income'] > 0) & (data['total_income'] <=100000)]['total_income'].count())
print('От 100 до 140 тыс.:', data[(data['total_income'] > 100000) & (data['total_income'] <=140000)]['total_income'].count())
print('От 140 до 190 тыс.:', data[(data['total_income'] > 140000) & (data['total_income'] <=190000)]['total_income'].count())
print('От 190 до 330 тыс.:', data[(data['total_income'] > 190000) & (data['total_income'] <=330000)]['total_income'].count())
print('От 330 тыс. и выше:', data[data['total_income'] > 330000]['total_income'].count())
print('Пропущенные данные:', data[data['total_income'] == -1]['total_income'].count())
print()

# Категоризируем данные по столбцу "ежемесячный доход" и запишем результат в новый столбец 
data['total_income_category'] = data['total_income'].apply(total_income_cat)    # Применение метода apply() к столбцу total_income с функцией total_income

# Выведем суммарные значения по новому столбцу total_income_category на экран для оценки
print('Результат работы функции категоризации по ежемесячному доходу:')
print(data['total_income_category'].value_counts())
```

- Есть ли зависимость между наличием детей и возвратом кредита в срок?

```python
# Функция для подготовки сводной таблицы для отчёта, в качестве аргумента задается название столбца с категориями
def data_pivot_fun(column_name):    
    # Обработаем данные с помощью метода pivot
    # Создание сводной таблицы из столбцов <column_name>, debt, с подсчётом количества значений функцией count 
    data_pivot = data.pivot_table(index=column_name, columns='debt', values='family_status_id', aggfunc='count')

    # Изменение названий столбцов для удобства восприятия итоговой таблицы методом set_axis
    data_pivot.set_axis(['нет долгов','есть долги'],axis = 'columns',inplace = True)

    # Добавление в сводную таблицу столбца с подсчётом общего количества клиентов
    data_pivot['всего клиентов'] = data_pivot['нет долгов'] + data_pivot['есть долги']

    # Добавление в сводную таблицу столбца с расчётом отношения должников к общему количеству клиентов с применением форматирования методом apply()
    try:
        data_pivot['% должников'] = (data_pivot['есть долги'] / data_pivot['всего клиентов']).apply('{:.2%}'.format) 
    except:
        return 'Не удалось вычислить отношение, попробуйте запустить программу заново'

    # Вывод сводной таблицы на экран с сортировкой по столбцу "% должников" по-убыванию
    return data_pivot.sort_values(by = '% должников', ascending = False)

# Вызов функции data_pivot_fun с аргументом children_category
display(data_pivot_fun('children_category'))
```
